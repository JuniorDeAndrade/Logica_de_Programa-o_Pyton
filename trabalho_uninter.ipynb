#Exerccio 1:

Escreva um programa que leia o nome de um aluno e sua nota final. Em seguida, informe o conceito conforme a tabela abaixo. A sada do programa deve exibir na tela uma frase com o padro descrito a seguir:

    De 7 a 8,9    |    B

# Resposta:

```python
"""
Programa para cadastra alunos e suas notas finais, e retorna o seu conceito
referente a sua nota...

"""
while True:
  aluno = input('Digite seu nome: ')
  nota_final = float(input('Digite sua nota final: '))
  # Separandoconjuntodecdigos,paramelhorvisualizaodousurio
  print('-' * 78) 
  conceito = 0
  # verificando notas, e atribuindo ao seu conceito...
  if nota_final >= 0 and nota_final < 3.0:
      conceito = 'E'

  elif nota_final < 5.0:
      conceito = 'D'

  elif nota_final < 7.0:
      conceito = 'C'

  elif nota_final < 9.00:
      conceito = 'B'

  elif nota_final <= 10.00:
      conceito = 'A'
  else:
    print('Erro: Digite um nota de (0)  (10)')
    continue
  print(f'Nome do Aluno: {aluno.title()}') # Formatandoparaasletrasiniciaismaisculas 
  print(f'Nota Final: {nota_final}')
  # Separandoconjuntodecdigos,paramelhorvisualizaodousurio
  print('-' * 78)
  print(f'O aluno {aluno.title()} tirou nota {nota_final} e se enquadra no conceito {conceito} ')
  print(' ')
  sair = input('Deseja digitar novos dados? (S): Sim ou (N): No:  ')
  # usando lower, caso usario digite a letra minscula.
  if sair.lower() == 'n':
    print('Programa encerrado...')
    break
  elif sair.lower() == 's':
    continue
  else:
    print('Erro: Digite (S) ou (N). Programa foi encerrado...')
    break

  
 
```

#Exerccio 3:

Um canal de jogos do youtube est fazendo um sorteio para angariar doaes para pessoas em situao de vulnerabilidade social. 

A cada 10,00 doado o nome da pessoa  inserido em uma lista de sorteio, por exemplo:

Ruth doou 20,00;
Maria doou 30,00;
Fernando doou 50,00;

A lista de sorteio estar com os valores:

listaSorteio = [Ruth, Ruth, Maria, Maria, Maria, Fernando, Fernando, Fernando, Fernando, Fernando]

Implemente um programa para cadastrar o nome das pessoas que doaram.

O programa deve embaralhar a lista, sortear o ganhador e imprimir o seu nome.

Imprima na tela um teste do seu programa utilizando como primeiro doador o seu nome e os dois ltimos dgitos do seu RU para o valor doado. 

No se esquea de imprimir tambm a lista de sorteio.

# Resposta:

```python
# Importandobibliotecanecessria...
from random import choice,shuffle

"""
Programa que cadastra doadores para uma promoo social...
A cada 10,00 reais doados o doador participa de um sorteio.
O programa conta quantas chances o doador ter; sorteia e mostra o vencedor.
"""

# imprimindo informaes.
print('Recebendo doaes para pessoas com vulnerabilidade social ')
print('Doe,acada10,00reaisdoadosvocganhauma oportunidadedeconcorreraumsorteio. ')
# Separandoconjuntodecdigos,paramelhorvisualizaodousurio
print('-' * 85)
print('Cadastrando Doadores. ')
participantes = []
# Tratandoerro,casoousuriosepareacasadecimaldovalorcom vrgula.
while True:
    try:
        nome = input('Digite um nome: ')
        valor = float(input('Quanto essa pessoa vai doar? '))
        if valor >= 10:
            vezes = valor / 10
            for i in range(int(vezes)):
                participantes.append(nome.title())
    except ValueError:
        print('Erro: Caso haja um valor com centavos, separe-os com Ponto "." ')  
    sair = input('Deseja sair: S/N: ')
    if sair.lower() == 's':
        break
    else:
        continue  
print(f'A lista de participantes: {participantes}')
# Separandoconjuntodecdigos,paramelhorvisualizaodousurio
print('-' * 78)
# Embaralhando a lista de participantes.
shuffle(participantes)

# Sorteando os participantes...
sorteio = choice(participantes)
print('-' * 78)
print(f'O grande vencedor do sorteio  : {sorteio} !!! ')

```

#Exerccio 4:

Considere a tabela a seguir referente a produtos armazenados em um depsito, em que so considerados o estoque atual de cada produto e o estoque mnimo necessrio

    Cdigo | Estoque | Mnimo |

    1      |   35    |  20    |
    5      |   75    |  50    |
    2      |   43    |  45    |
    3      |   26    |  18    |
    20     |   35    |  20    |

Armazene as informaes acima em uma estrutura de lista com dicionrio, substituindo a primeira linha com os dados: 
no campo cdigo coloque o primeiro digito do seu RU, no estoque os dois dgitos seguintes de seu RU, e no campo mnimo os dois ltimos dgitos do seu RU.

Por exemplo, tendo o RU: 123456

As informaes devem ser inseridas no dicionrio via teclado. 

Ao digitar o cdigo 0 (zero), o programa interrompe a leitura e se encerra.

 Ordene os produtos em ordem crescente de cdigo. 
 
 Imprima na tela um teste do seu programa usando como primeiro cadastro o primeiro digito do seu RU, como estoque os dois dgitos seguintes de seu RU, e como mnimo os dois ltimos dgitos do seu RU.

# Resposta:

```python
"""
Programaquecadastraoid,estoqueeestoquemnimodeprodutos.
Osdadossoarmazenadosemumalista,separadopordicionrios.
"""
estoque = []
while True:
    # Recebendo dados de entrada do usurio.
    codigo_produto = int(input('Digite o cdigo do produto: '))
    if codigo_produto <= 0: # condio para encerramento do programa.
        break
    else:
        quantidade = int(input('Digite a quantidade desse produto: '))
        estoque_min = int(input('Digite a quantidade minima para esse produto: '))
        dados = {'codigo': codigo_produto, 'estogue': quantidade, 'minimo': estoque_min}
        estoque.append(dados)
        ordenar = sorted(estoque, key=lambda dados: dados['codigo'])
print(ordenar)
```
